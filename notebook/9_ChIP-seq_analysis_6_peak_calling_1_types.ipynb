### 9.6.1 ChIP-seq データの種類
結合タンパク質の種類によって、3つのクラスに分類できます。

1. **Sharp(point signal):** 数 100bp 程度の短いゲノム領域からなるピーク。主に、転写因子や翻訳後のヒストン修飾(H3K4me3 など)が該当する。
1. **Broad(wide signal):** 数 1000bp に渡る広い範囲のゲノム領域からなるピーク。主に、gene body に分布する H3K36me3、ポリコーム複合体に分布する H3K23me3などが該当する。
1. **Mixed:** 上記のピークが混在するケース。RNA Polymerase 2 がそれに該当する。

ピークの特性に合わせて解析ツールも選択する必要がある。

ツールの選択を見極める最も有効な手段は、データを可視化することである。ここでは、R を使って、H3K4me3 (Sharp), H3K36me3 (Broad), and POL2 (Mixed) ChIP-seq サンプルをそれぞれ可視化してみる。

```r
# ChIP-seq のデータのルートパス
data_path = "../data/chip-seq"

# ChIP-seq bigWig ファイルのリスト
chip_files = list(
    H3K4me3  = 'GM12878_hg38_H3K4me3.chr21.bw',

    H3K36me3 = 'GM12878_hg38_H3K36me3.chr21.bw',

    POL2     = 'GM12878_hg38_POLR2A.chr21.bw'
)

# 各ファイルのフルパスを取得
chip_files = lapply(chip_files, function(x){
    file.path(data_path, x)
})
chip_files
```

```r
# rtracklayer ライブラリの読み込み
library(rtracklayer)
```

```r
# ChIP bigWig ファイルの読み込み
chip_profiles = lapply(chip_files, rtracklayer::import.bw)
```

```r
# AnnotationHub ライブラリの読み込み
library(AnnotationHub)
```

```r
# ヒト遺伝子のアノテーション情報を取得(ダウンロード)
# MEMO: GRCh38.92 -> AH61126 のデータを使用
# MEMO: 読み込みにかなり時間がかかる(5minぐらい)
hub = AnnotationHub()
gtf = hub[['AH61126']]
```

```r
# chr21 のみを抽出
seqlevels(gtf, pruning.mode='coarse') = '21'

# Ensembl のデータは染色体名が「1, 2, 3,...」となっているので、「chr1, chr2, chr3,...」の形式に変換する
# 染色体名を抽出
ensembl_seqlevels = seqlevels(gtf)

# 染色体名に chr の prefix をつける
ucsc_seqlevels = paste0('chr', ensembl_seqlevels)

# 染色体名を置換
seqlevels(gtf, pruning.mode='coarse') = ucsc_seqlevels
```

```r
gtf
```

遺伝子のアノテーション情報と一緒に ChIP-seq データを可視化したいので、GRanges オブジェクトを transcript のデータベース(data.base データ構造)に変換する。

```r
# GenomicFeatures ライブラリの読み込み
library(GenomicFeatures)
```

```r
# gtf アノテーション情報を data.base オブジェクトに変換
txdb = makeTxDbFromGRanges(gtf)
txdb
```

```r
# Gviz ライブラリの読み込み
# MEMO: テキストが間違っており、Gviz ライブラリの読み込みを先に行う必要がある(GeneRegionTrack 関数が必要なため)。
library(Gviz)
```

```r
# transcript データベースを gene track オブジェクトとして定義
gene_track = GeneRegionTrack(txdb, chr='chr21', genome='hg38')
```

```r
gene_track
```

```r
# ヒトゲノムの染色体長を取得
# getChromInfoFromUCSC： UCSC のゲノム情報を使用する
hg_chrs = getChromInfoFromUCSC('hg38')

# ヒトゲノムの21番染色体を取得
# grepl: chr21$ に一致するかどうか boolean で返す
# subset: data.frame(データフレーム) から grepl で true となる対象データを抽出する 
hg_chrs = subset(hg_chrs, grepl('chr21$',chrom))
hg_chrs
```

```r
# ダウンロードしたヒトゲノムの染色体長のデータ: data.frame(データフレーム)というデータ構造
# これを data.frame(データフレーム) から 名前付きベクター(配列)にデータ構造を変換する
# setnames: chrom を名前とした size の名前付きベクターを作成する
# with: データ構造を変換したり、データの絞り込みを行ったりできる(ここでは名前付きベクターにデータ構造を変換している)
seqlengths = with(hg_chrs, setNames(size, chrom))
seqlengths
```

```r
# ideogram トラックを構築
# MEMO: 染色体の模式図(イデオグラム)のトラックを作成している
chr_track   = IdeogramTrack(
    chromosome = 'chr21', 
    genome     = 'hg38'
)
```

```r
# ゲノムの座標軸のトラックを構築
axis = GenomeAxisTrack(
    range = GRanges('chr21', IRanges(1, width=seqlengths))
)
```

```r
chip_profiles
```

```r
# bw ファイルのカバレッジデータを 1ファイルずつ data トラックとして定義しリストにする
# MEMO: chip_profiles の中身は上述の通り
# MEMO: https://www.bioconductor.org/packages/devel/bioc/vignettes/Gviz/inst/doc/Gviz.html#43_DataTrack
data_tracks = lapply(names(chip_profiles), function(exp_name){
    
    # chip_profiles[[exp_name]] - 指定のデータ exp_name のカバレッジデータをセットする
    DataTrack(
        range = chip_profiles[[exp_name]],   
        name  = exp_name,  
        
        # トラックのタイプ(ヒストグラムのライン)
        type  = 'h', 
        
        # ライン幅のパラメータ
        lwd   = 5
    )
})
```

GViz ライブラリのドキュメントから抜粋。


```r
# URB1 遺伝子の開始位置を選択
start = min(start(subset(gtf, gene_name == 'URB1')))

# URB1 遺伝子の終了位置を選択
end   = max(end(subset(gtf, gene_name == 'URB1')))
```

```r
# URB1 遺伝子の前後 5000bp を描画
plotTracks(
    trackList = c(chr_track, axis, gene_track, data_tracks),
    
    # 相対的なトラックサイズ
    sizes     = c(1,1,1,1,1,1), 
    
    # タイトルのバックグラウンドのカラー
    background.title     = "black",
    
    # 遺伝子セットの可視化の制御
    collapseTranscripts  = "longest", 
    transcriptAnnotation = "symbol",
    
    # 可視化範囲 
    from = start - 5000,
    to   = end   + 5000
)
```

- **H3K36me3:** gene body 全体に enrich しているのが確認できる。
- **Pol2**: TSS 領域に sharp なピーク、gene body 全体にも enrich しているのが確認できる(mixed)。
