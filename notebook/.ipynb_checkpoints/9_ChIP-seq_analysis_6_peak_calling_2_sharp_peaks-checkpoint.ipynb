## 9.6 Peak calling
### 9.6.2 Peak calling: Sharp peaks
sharp / broad ピークの両方を検出するために `normR` ライブラリをここでは使用する。

```r
# ChIP-seq のデータのルートパス
data_path = "../data/chip-seq"

# ChIP データのファイル
chip_file    = file.path(data_path, 'GM12878_hg38_CTCF_r1.chr21.bam')

# Control データのファイル
control_file = file.path(data_path, 'GM12878_hg38_Input_r5.chr21.bam')
```

```r
library(GenomicRanges)
library(GenomicAlignments)
```

```r
# ヒトゲノムの染色体長を取得
# getChromInfoFromUCSC： UCSC のゲノム情報を使用する
hg_chrs = getChromInfoFromUCSC('hg38')

# ヒトゲノムの21番染色体を取得
# grepl: chr21$ に一致するかどうか boolean で返す
# subset: data.frame(データフレーム) から grepl で true となる対象データを抽出する 
hg_chrs = subset(hg_chrs, grepl('chr21$',chrom))
hg_chrs
```

```r
# ダウンロードしたヒトゲノムの染色体長のデータ: data.frame(データフレーム)というデータ構造
# これを data.frame(データフレーム) から 名前付きベクター(配列)にデータ構造を変換する
# setnames: chrom を名前とした size の名前付きベクターを作成する
# with: データ構造を変換したり、データの絞り込みを行ったりできる(ここでは名前付きベクターにデータ構造を変換している)
seqlengths = with(hg_chrs, setNames(size, chrom))
seqlengths
```

```r
# 1kb windows を定義
tilling_window = unlist(tileGenome(seqlengths, tilewidth=1000))
tilling_window
```

```r
# リードカウント(ChIP, Control)
counts         = summarizeOverlaps(
    features = tilling_window, 
    reads    = c(chip_file, control_file)
)
counts
```

```r
# リード数をノーマライズ(CPM を計算)
counts         = assays(counts)[[1]]
cpm = t(t(counts)*(1000000/colSums(counts)))
```

```r
library(ggplot2)
```

```r
# matrix を data.frame にデータ変換
cpm = data.frame(cpm)
```

```r
# データの可視化(ダイナミックレンジの enrichment の散布図)
ggplot(
    data = cpm, 
    aes(
        x = GM12878_hg38_Input_r5.chr21.bam, 
        y = GM12878_hg38_CTCF_r1.chr21.bam)
    ) +
    geom_point(alpha = 0.2) +
    # 傾きが 1 の直線を引く
    geom_abline(slope = 1) +
    # 濃いめの枠線をつける
    theme_bw() +
    # 色パレット
    scale_fill_brewer(palette='Set2') +
    theme(
        axis.text   = element_text(size=10, face='bold'),
        axis.title  = element_text(size=14,face="bold"),
        plot.title  = element_text(hjust = 0.5),
        axis.text.x = element_text(hjust = 0.5)) +
    xlab('Input CPM') +
    ylab('CTCF CPM') +
    ggtitle('ChIP versus Input')
```

線より上の点は、ChIP サンプルのほうが enrich している、  
線より下の点は、Input サンプルのほうが enrich している、  
ということを示す。

以下では、`normR` ライブラリを用いて、統計的に有意に ChIP サンプルで enrich しているピークを同定する。

```r
library(normr)
```

```r
# Chip / Control サンプルを使って peak calling
ctcf_fit = enrichR(
    
            # ChIP ファイル
            treatment = chip_file,
            
            # Control ファイル
            control   = control_file,
            
            # ヒトゲノムのバージョン
            genome    = "hg38",
            
            # 途中経過のログを出力するかどうか
            verbose   = FALSE)
```

```r
# Summary を見てみる
summary(ctcf_fit)
```

q-value < 0.1 であるピークが、(627 + 120 + 195 + 87) = 1029 ある。  
Background が 97.72%。

```r
# ピーク領域を GRanges オブジェクトとして取得
ctcf_peaks = getRanges(ctcf_fit)
```

```r
# 各ピーク領域に対応する q-value を取得して、GRanges オブジェクトの列に追加
ctcf_peaks$qvalue = getQvalues(ctcf_fit)
```

```r
# 計算した enrichment を取得して、GRanges オブジェクトの列に追加
ctcf_peaks$enrichment = getEnrichment(ctcf_fit)
```

```r
# リード数が 0 カウントの領域をカットする(component 列が NA の領域をカット)
ctcf_peaks = subset(ctcf_peaks, !is.na(component))
```

```r
# q value < 0.01 のデータのみ抽出
ctcf_peaks = subset(ctcf_peaks, qvalue < 0.01)
```

```r
# q value が小さい順に並び替え
ctcf_peaks = ctcf_peaks[order(ctcf_peaks$qvalue)]
ctcf_peaks
```

```r
# chr21 のデータのみ抽出(サンプルデータ削減のため)
seqlevels(ctcf_peaks, pruning.mode='coarse') = 'chr21'
```

```r
# txt ファイルとしてファイル出力
write.table(ctcf_peaks, file.path(data_path, 'CTCF_peaks.txt'), 
            row.names=F, col.names=T, quote=F, sep='\t')
```

```r
# tilling_window と overlap するピーク領域を特定し、boolean 型で返す
# 同定されたピーク領域の場合、true
# ピーク領域でない場合、false
enriched_regions = countOverlaps(tilling_window, ctcf_peaks) > 0
```

```r
library(ggplot2)
```

```r
# cpm に enriched_regions 列を追加
cpm$enriched_regions = enriched_regions 
```

```r
# MEMO: 赤プロット(同定されたピーク領域のプロット)を前面にしたいのでソート
sorted_cpm = cpm[order(cpm$enriched_regions, decreasing=F),]
```

```r
ggplot(
    data = sorted_cpm, 
    aes(
        x = GM12878_hg38_Input_r5.chr21.bam, 
        y = GM12878_hg38_CTCF_r1.chr21.bam, 
        color = enriched_regions
    )) +
  geom_point(alpha = 0.3) +
  geom_abline(slope = 1) +
  theme_bw() +
  scale_fill_brewer(palette='Set2') +
  theme(
    axis.text   = element_text(size=10, face='bold'),
    axis.title  = element_text(size=14,face="bold"),
    plot.title  = element_text(hjust = 0.5),
    axis.text.x = element_text(hjust = 0.5)) +
  xlab('Input CPM') +
  ylab('CTCF CPM') +
  ggtitle('ChIP versus Input') +
  scale_color_manual(values=c('gray','red'))
```

Input サンプルのほうが enrich しているにもかかわらず、`normR` で統計的に有意差がある領域が同定されている(後で確認)。  
続いて、ピーク領域周辺のカバレッジデータを可視化してみる。

```r
# 1つの BAM ファイルから、カバレッジデータを計算する関数
calculateCoverage = function(
  bam_file,
  extend = 200
){
  
  # BAM ファイルからリードデータを取得
  reads = readGAlignments(bam_file)
  
  # GRanges オブジェクトに変換
  reads = granges(reads)
  
  # リード長を 200bp に補正
  reads = resize(reads, width=extend, fix='start')
  
  # カバレッジ vector を取得
  cov   = coverage(reads)
  
  # sequencing depth に基づき補正(CPM を計算 & 小数点第２位以下を丸める)
  cov = round(cov * (1000000/length(reads)),2)
  
  # GRanges オブジェクトに変換
  cov   = as(cov, 'GRanges')
  
  # chr21 のみ抽出
  seqlevels(cov, pruning.mode='coarse') = 'chr21'
  return(cov)
}
```

```r
# ChIP ファイルのカバレッジを計算
ctcf_cov = calculateCoverage(chip_file)

# Control ファイルのカバレッジを計算
cont_cov = calculateCoverage(control_file)
```

```r
library(Gviz)
```

```r
# ideogram トラックを構築
# MEMO: 染色体の模式図(イデオグラム)のトラックを作成している
chr_track   = IdeogramTrack(
    chromosome = 'chr21', 
    genome     = 'hg38'
)
```

```r
# ゲノムの座標軸のトラックを構築
axis = GenomeAxisTrack(
    range = GRanges('chr21', IRanges(1, width=seqlengths))
)
```

```r
# peaks トラック
peaks_track = AnnotationTrack(ctcf_peaks, name = "CTCF Peaks")
```

```r
# カバレッジのトラック
chip_track  = DataTrack(
    range = ctcf_cov,   
    name  = "CTCF",  
    type  = 'h', 
    lwd   = 3
)

cont_track  = DataTrack(
    range = cont_cov,   
    name  = "Input", 
    type  = 'h', 
    lwd   = 3
)
```
